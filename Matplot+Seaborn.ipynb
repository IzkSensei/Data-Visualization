```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt

import sklearn
from sklearn.linear_model import LinearRegression


import seaborn as sns
```

## Seaborn datasets

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

## Scatter plot

```python
plt.style.use('seaborn')
```

```python
sns.pairplot(tips, hue="day")
```

```python
ax = sns.scatterplot(x="total_bill", y="tip",c=tips["size"],cmap='viridis', data=tips)
```

```python
ax = sns.scatterplot(x="total_bill", y="tip",hue="sex",size=tips["size"],style='time',cmap='viridis', data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter",
            hue="sex", size="size", 
            col='day', row='time', style='time',
            col_order=["Thur", "Fri", "Sat", 'Sun']
            )
```

```python
sns.catplot(x="sex", y="total_bill", data=tips,hue="time",col="day",col_order=["Thur","Fri","Sat","Sun"])
```

```python
sns.swarmplot(x ="size", y = "total_bill",hue="day", data = tips);
```

```python
sns.boxplot(x="sex", y="total_bill", data=tips, color=".8")
sns.swarmplot(x="sex", y="total_bill", hue="day", data=tips)
plt.gcf().set_size_inches(11, 8)
```

```python
jp = sns.jointplot("total_bill", "tip", data=tips, kind="reg")
```

```python
jp = sns.jointplot("total_bill", "tip", data=tips, kind="kde")
```

```python
dp = sns.displot(data=tips, x="total_bill", row="sex", col="day",hue="day", kind="hist")
```
